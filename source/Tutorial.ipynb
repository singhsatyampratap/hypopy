# Downloading the Station Data

```python
from obspy import UTCDateTime
from obspy.clients.fdsn import Client

client = Client("IRIS")
t = UTCDateTime("2011-03-11T05:46:23")  # Tohoku
inv = client.get_stations(network="II",station= "*",channel="*", starttime=t + 10 * 60, endtime=t + 30 * 60)
print(inv)
#st.plot()
```

```python
network=inv[0]
n=len(network)
print("Total number of station in Network= {}".format(n))
```

```python
network_name="II"
filename="Stationlist"
for i in range(n):
    hypopy.add_station(filename,network[i].code,network_name,"BHZ",network[i].latitude,network[i].longitude,network[i].elevation)
    hypopy.add_station(filename,network[i].code,network_name,"BH1",network[i].latitude,network[i].longitude,network[i].elevation)
    hypopy.add_station(filename,network[i].code,network_name,"BH2",network[i].latitude,network[i].longitude,network[i].elevation)
```
